```python
import sys
import pandas as pd
```

```python
regions = pd.read_csv("regions_to_remove.txt",index_col=None,header=0,sep="\t")
```

```python
regions
```

```python
data = pd.read_csv("../../big_dataframe_Xist.txt",sep="\t",index_col=0,header=0)
normdata = pd.read_csv("../../big_dataframe_Xist_diffBUM_HMM_scaled_dc.txt",sep="\t",index_col=0,header=0)
```

```python
data.head()
```

```python
for region in regions.index:
    start,end = regions.loc[region].values
    start -= 1 ### convert to zero-based data
    data.drop(data.index[start:end])
    normdata.drop(normdata.index[start:end])
```

```python
data.to_csv("../../big_dataframe_Xist_regions_removed.txt",sep="\t")
normdata.to_csv("../../big_dataframe_Xist_diffBUM_HMM_scaled_dc_regions_removed.txt",sep="\t")
```
