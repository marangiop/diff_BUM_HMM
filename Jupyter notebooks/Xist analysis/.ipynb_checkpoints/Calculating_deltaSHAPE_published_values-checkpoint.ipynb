```python
import sys
import numpy as np
import pandas as pd
```

```python
## Loading the data
deltashaperepone = pd.read_csv('../../deltaSHAPE_confirmed_sites/deltaSHAPE_rep1.igv',sep="\t")
deltashapereptwo = pd.read_csv('../../deltaSHAPE_confirmed_sites/deltaSHAPE_rep2.igv',sep="\t")
alldata  = pd.read_csv('../../big_dataframe_Xist_diffBUM_HMM_scaled_dc.txt',sep="\t",header=0,index_col=0)
previous = pd.read_csv('../../big_dataframe_Xist.txt',sep="\t",header=0,index_col=0)
```

### Merging the two datasets and only keeping those that had deltaSHAPE values in both

```python
merged = deltashaperepone.merge(deltashapereptwo)
```

```python
merged.head()
```

```python
col = merged[["deltaSHAPE","deltaSHAPE_rep1"]]
merged['average'] = col.mean(axis=1)
```

```python
merged.head()
```

```python
merged.to_csv("../../deltaSHAPE_confirmed_sites/deltaSHAPE_merged_averaged.txt",sep="\t")
```

### Adding the new information to the big table:

```python
alldata[alldata.columns[2:5]] = 0
previous[previous.columns[2:5]] = 0
```

```python
subset = alldata[alldata.nucleotide_number.isin(merged.Start.values)].index
```

```python
alldata.loc[subset,"average_deltaSHAPE"] = merged.average.values
```

```python
alldata.loc[subset].head()
```

```python
alldata.head()
```

```python
exvivoindex = alldata[alldata.average_deltaSHAPE > 1].index
invivoindex = alldata[alldata.average_deltaSHAPE < 1].index
```

```python
alldata.loc[exvivoindex,"DeltaSHAPE_ex_vivo"] = 1
alldata.loc[invivoindex,"DeltaSHAPE_in_vivo"] = 1
```

```python
previous.loc[exvivoindex,"DeltaSHAPE_ex_vivo"] = 1
previous.loc[invivoindex,"DeltaSHAPE_in_vivo"] = 1
```

```python
alldata.to_csv('../../big_dataframe_Xist_diffBUM_HMM_scaled_dc_deltaSHAPE_corrected.txt',sep="\t")
previous.to_csv('../../big_dataframe_Xist_deltaSHAPE_corrected.txt',sep="\t")
```
