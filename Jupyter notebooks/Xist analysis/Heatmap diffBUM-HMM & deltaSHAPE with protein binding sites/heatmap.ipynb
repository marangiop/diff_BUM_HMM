```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
flights = sns.load_dataset("flights")
```

```python
flights = flights.pivot("month", "year", "passengers")
ax = sns.heatmap(flights)
```

```python
df = pd.DataFrame(index=np.arange(6), columns=['TF','diffBUM_HMM_exvivo', 'diffNBUM_HMM_incell','deltaSHAPE_exvivo', 'deltaSHAPE_incell' ])         
tf=['CELF1','FUS', 'HuR', 'PTBP1', 'RBFOX2', 'TARDBP']
exvivo_diffbumhmm = [68,411,104,59,3, 65]  
incell_diffbumhmm  = [16,46, 20, 15, 2,2] 

exvivo_deltashape = [104, 162, 171, 95, 3, 8]  
incell_deltashape = [24, 192, 34, 21, 0, 16] 


df.loc[:,'TF'] = tf
df.loc[:,'diffBUM_HMM_exvivo'] = exvivo_diffbumhmm  
df.loc[:,'diffNBUM_HMM_incell'] = incell_diffbumhmm  

df.loc[:,'deltaSHAPE_exvivo'] = exvivo_deltashape
df.loc[:,'deltaSHAPE_incell'] = incell_deltashape
df=df.set_index('TF')


#import matplotlib.pyplot as plt
print(df)
#fig, ax = plt.subplots(figsize=(8,5))  
ax = sns.heatmap(df, linewidths=.5, annot=True, fmt="d", cbar_kws={'ticks': [0.0, 40, 80, 120, 160, 200, 240, 280, 320, 360,400]})
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')

```

```python
ax = sns.heatmap(df, cmap="YlGnBu", linewidths=.5, annot=True, fmt="d", cbar_kws={'ticks': [0.0, 40, 80, 120, 160, 200, 240, 280, 320, 360,400]})
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')
```
