```python
import seaborn as sns
import pandas as pd
import numpy as np
import itertools
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['Arial']
rcParams['pdf.fonttype'] = 42
```

```python
threshold = 0.95
```

```python
data = pd.read_csv("../../../big_dataframe_Xist_diffBUM_HMM_scaled_dc_deltaSHAPE_corrected.txt",\
                  sep="\t",\
                  header=0,\
                  index_col=0)
previous = pd.read_csv("../../../big_dataframe_Xist_deltaSHAPE_corrected.txt",\
                  sep="\t",\
                  header=0,\
                  index_col=0)
```

```python
data.head()
```

```python
previous.head()
```

```python
proteins = ['CELF1','FUS', 'HuR', 'PTBP1', 'RBFOX2', 'TARDBP']
samples  = ['diffBUM_HMM_exvivo_norm','diffBUM_HMM_exvivo','diffBUM_HMM_incell_norm','diffBUM_HMM_incell',
            'deltaSHAPE_exvivo_1','deltaSHAPE_exvivo_2','deltaSHAPE_incell_1','deltaSHAPE_incell_2']

df = pd.DataFrame(0,index=proteins,columns=samples)
```

```python
for protein in proteins:
    df.loc[protein,'diffBUM_HMM_exvivo_norm'] = data[data.diffBUMHMM_ex_vivo > threshold][protein].sum()
    df.loc[protein,'diffBUM_HMM_exvivo'] = previous[previous.diffBUMHMM_ex_vivo > threshold][protein].sum()
    df.loc[protein,'diffBUM_HMM_incell_norm'] = data[data.diffBUMHMM_in_cell > threshold][protein].sum()
    df.loc[protein,'diffBUM_HMM_incell'] = previous[previous.diffBUMHMM_in_cell > threshold][protein].sum()
    df.loc[protein,'deltaSHAPE_exvivo_1'] = data[data.deltaSHAPE_rep1 > 0][protein].sum()
    df.loc[protein,'deltaSHAPE_exvivo_2'] = data[data.deltaSHAPE_rep2 > 0][protein].sum()
    df.loc[protein,'deltaSHAPE_incell_1'] = data[data.deltaSHAPE_rep1 < 0][protein].sum()
    df.loc[protein,'deltaSHAPE_incell_2'] = data[data.deltaSHAPE_rep2 < 0][protein].sum()
```

```python
df
```

```python
sns.heatmap(df,\
            cmap='Blues',\
            linewidths=.5,\
            annot=True,\
            fmt="d",\
            cbar_kws={'ticks': [0.0, 40, 80, 120, 160, 200, 240, 280, 320, 360, 400]})

plt.savefig("heatmap_blues_diffbumhmm_vs_deltashape_TF_sites.pdf")
```
