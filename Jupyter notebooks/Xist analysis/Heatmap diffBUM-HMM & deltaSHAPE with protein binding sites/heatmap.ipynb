```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
df = pd.DataFrame(index=np.arange(6),columns=['TF','diffBUM_HMM_exvivo_norm', 'diffNBUM_HMM_incell_norm','diffBUM_HMM_exvivo', 'diffNBUM_HMM_incell','deltaSHAPE_exvivo', 'deltaSHAPE_incell' ])         
tf=['CELF1','FUS', 'HuR', 'PTBP1', 'RBFOX2', 'TARDBP']

#Threshold used for diffBUM-HMM output is 0.95
exvivo_diffbumhmm_normalised = [60,375, 96, 53, 2, 58]  
incell_diffbumhmm_normalised  = [11,33, 17, 10, 2,2] 

exvivo_diffbumhmm_NOT_normalised = [40,268, 56, 37, 1, 38]  
incell_diffbumhmm_NOT_normalised  = [13,26, 12, 11, 1,4] 

exvivo_deltashape = [104, 162, 171, 95, 3, 8]  
incell_deltashape = [24, 192, 34, 21, 0, 16] 


df.loc[:,'TF'] = tf

df.loc[:,'diffBUM_HMM_exvivo_norm'] = exvivo_diffbumhmm_normalised  
df.loc[:,'diffNBUM_HMM_incell_norm'] = incell_diffbumhmm_normalised  

df.loc[:,'diffBUM_HMM_exvivo'] = exvivo_diffbumhmm_NOT_normalised  
df.loc[:,'diffNBUM_HMM_incell'] = incell_diffbumhmm_NOT_normalised    

df.loc[:,'deltaSHAPE_exvivo'] = exvivo_deltashape
df.loc[:,'deltaSHAPE_incell'] = incell_deltashape
df=df.set_index('TF')


#import matplotlib.pyplot as plt
print(df)
#fig, ax = plt.subplots(figsize=(8,5))  
ax = sns.heatmap(df, linewidths=.5, annot=True, fmt="d", cbar_kws={'ticks': [0.0, 40, 80, 120, 160, 200, 240, 280, 320, 360,400]})
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')

```

```python
ax = sns.heatmap(df, cmap="Blues", linewidths=.5, annot=True, fmt="d", cbar_kws={'ticks': [0.0, 40, 80, 120, 160, 200, 240, 280, 320, 360,400]})
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')

matplotlib.rcParams['font.family'] = 'Arial'
matplotlib.rcParams['pdf.fonttype'] = 42
plt.tight_layout()
plt.savefig("heatmap_blues_diffbumhmm_vs_deltashape_TF_sites.pdf")
```
