```python
#ADDING DELTA SHAPE INFORMATION
#EX VIVO

df_5 = pd.read_csv(path + 'Data/Map files Xist/XIST_exvivo_1M7.txt', sep="\t", header=0)
df_5.head

#Create empty column
df["DeltaSHAPE_ex_vivo"] = 0

new_list=[]
for index, row in df.iterrows():
    new_list.append(row['nucleotide_number'])
#print(new_list)

new_dict = dict(zip(df_5.Nuc, df_5.DeltaSHAPE))
#print(new_dict)

for key, value in new_dict.items():
    if key in new_list:
        rowIndex = df.loc[df['nucleotide_number']==key].index
        df.loc[rowIndex, 'DeltaSHAPE_ex_vivo'] = value
        
#df.to_csv('cavolo.txt', sep='\t')

#ADDING DELTA SHAPE INFORMATION
#IN CELL
df_6 = pd.read_csv(path + 'Data/Map files Xist/XIST_incell_1M7.txt', sep="\t", header=0)
df_6.head

#Create empty column
df["DeltaSHAPE_in_cell"] = 0

new_list_2=[]
for index, row in df.iterrows():
    new_list_2.append(row['nucleotide_number'])


new_dict_2 = dict(zip(df_6.Nuc, df_6.DeltaSHAPE))


for key, value in new_dict_2.items():
    if key in new_list_2:
        rowIndex = df.loc[df['nucleotide_number']==key].index
        df.loc[rowIndex, 'DeltaSHAPE_ex_vivo'] = value
        


#df.to_csv('cavolo.txt', sep='\t')  
```

```python
#path="C:/Users/maran/Desktop/diff_BUM_HMM_Project/Github/diff_BUM_HMM/

file_list=os.listdir(path + "/Data/Bed files Xist/CLIPdb_transcript_coords/")

for file in file_list:
    
    colnames=['chrom', 'chromStart', 'chromEnd', 'name', 'score', 'strand', 'Do not know'] 
    new_df = pd.read_csv(path + "/Data/Bed files Xist/CLIPdb_transcript_coords/"+ file, sep="\t",names=colnames, header=None)
    
    if 'CELF1' in file:
        if 'CELF1_binding_site' not in list(df.columns.values):
            df["CELF1_binding_site"] = np.nan
        new_dict = dict(zip(new_df.chromStart, new_df.chromEnd))
        for key, value in new_dict.items(): 
            rowIndex = df.loc[df['nucleotide_number']==key].index
            df.loc[rowIndex, 'CELF1_binding_site'] = 1
            rowIndex = df.loc[df['nucleotide_number']==value].index
            df.loc[rowIndex, 'CELF1_binding_site'] = 1
            diff=value-key
            for i in range(diff):
                new_key = key+i
                rowIndex = df.loc[df['nucleotide_number']==new_key].index
                df.loc[rowIndex, 'CELF1_binding_site'] = 1   
    
    elif 'FUS' in file:
        if 'FUS_binding_site' not in list(df.columns.values):
            df["FUS_binding_site"] = np.nan
        new_dict = dict(zip(new_df.chromStart, new_df.chromEnd))
        for key, value in new_dict.items(): 
            rowIndex = df.loc[df['nucleotide_number']==key].index
            df.loc[rowIndex, 'FUS_binding_site'] = 1
            rowIndex = df.loc[df['nucleotide_number']==value].index
            df.loc[rowIndex, 'FUS_binding_site'] = 1
            diff=value-key
            for i in range(diff):
                new_key = key+i
                rowIndex = df.loc[df['nucleotide_number']==new_key].index
                df.loc[rowIndex, 'FUS_binding_site'] = 1     
    
    elif 'HuR' in file:
        if 'HuR_binding_site' not in list(df.columns.values):
            df["HuR_binding_site"] = np.nan
        new_dict = dict(zip(new_df.chromStart, new_df.chromEnd))
        for key, value in new_dict.items(): 
            rowIndex = df.loc[df['nucleotide_number']==key].index
            df.loc[rowIndex, 'HuR_binding_site'] = 1
            rowIndex = df.loc[df['nucleotide_number']==value].index
            df.loc[rowIndex, 'HuR_binding_site'] = 1
            diff=value-key
            for i in range(diff):
                new_key = key+i
                rowIndex = df.loc[df['nucleotide_number']==new_key].index
                df.loc[rowIndex, 'HuR_binding_site'] = 1      
    

    elif 'PTBP1' in file:
        if 'PTBP1_binding_site' not in list(df.columns.values):
            df["PTBP1_binding_site"] = np.nan
        new_dict = dict(zip(new_df.chromStart, new_df.chromEnd))
        for key, value in new_dict.items(): 
            rowIndex = df.loc[df['nucleotide_number']==key].index
            df.loc[rowIndex, 'PTBP1_binding_site'] = 1
            rowIndex = df.loc[df['nucleotide_number']==value].index
            df.loc[rowIndex, 'PTBP1_binding_site'] = 1
            diff=value-key
            for i in range(diff):
                new_key = key+i
                rowIndex = df.loc[df['nucleotide_number']==new_key].index
                df.loc[rowIndex, 'PTBP1_binding_site'] = 1    
   
    elif 'RBFOX2' in file:
        if 'RBFOX2_binding_site' not in list(df.columns.values):
            df["RBFOX2_binding_site"] = np.nan
        new_dict = dict(zip(new_df.chromStart, new_df.chromEnd))
        for key, value in new_dict.items(): 
            rowIndex = df.loc[df['nucleotide_number']==key].index
            df.loc[rowIndex, 'RBFOX2_binding_site'] = 1
            rowIndex = df.loc[df['nucleotide_number']==value].index
            df.loc[rowIndex, 'RBFOX2_binding_site'] = 1
            diff=value-key
            for i in range(diff):
                new_key = key+i
                rowIndex = df.loc[df['nucleotide_number']==new_key].index
                df.loc[rowIndex, 'RBFOX2_binding_site'] = 1
                
    elif 'TARDBP' in file:
        if 'TARDBP_binding_site' not in list(df.columns.values):
            df["TARDBP_binding_site"] = np.nan
        new_dict = dict(zip(new_df.chromStart, new_df.chromEnd))
        for key, value in new_dict.items(): 
            rowIndex = df.loc[df['nucleotide_number']==key].index
            df.loc[rowIndex, 'TARDBP_binding_site'] = 1
            rowIndex = df.loc[df['nucleotide_number']==value].index
            df.loc[rowIndex, 'TARDBP_binding_site'] = 1
            diff=value-key
            for i in range(diff):
                new_key = key+i
                rowIndex = df.loc[df['nucleotide_number']==new_key].index
                df.loc[rowIndex, 'TARDBP_binding_site'] = 1 

```
