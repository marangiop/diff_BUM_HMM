```python
import pandas as pd
import numpy as np
import os
import glob
```

```python
colnames=['nucleotide_number', 'SHAPE_reactivity', 'SE_reactivity', 'nucleotide_identity'] 
df = pd.read_csv("../../Data/Map files Xist/XIST_exvivo_1M7_rep1.map", sep="\t",names=colnames, header=None,index_col=0)
df.head()
```

```python
df = df.drop(df[["SHAPE_reactivity","SE_reactivity"]],axis=1)
```

```python
#Create empty column
df["DeltaSHAPE_ex_vivo"] = 0
df["DeltaSHAPE_in_cell"] = 0
df["DeltaSHAPE"] = 0
```

```python
df.head()
```

```python
df
```

```python
#ADDING DELTA SHAPE INFORMATION

deltashape = pd.read_csv('../../Data/Map files Xist/XIST_exvivo_vs_invivo_reps1both_1M7.txt', sep="\t", header=0, index_col=0)
#invivo = pd.read_csv('../../Data/Map files Xist/XIST_incell_1M7.txt', sep="\t", header=0, index_col=0)
```

```python
deltashape
```

```python


df["DeltaSHAPE"].loc[deltashape.index] = deltashape["DeltaSHAPE"]
```

```python
for index, row in df.iterrows():
    if row["DeltaSHAPE"] > 0:
        df.loc[row.name, 'DeltaSHAPE_ex_vivo'] = 1
        df.loc[row.name, 'DeltaSHAPE_in_cell'] = 0
        print('positive')
    elif row["DeltaSHAPE"] < 0:
        df.loc[row.name, 'DeltaSHAPE_ex_vivo'] = 0
        df.loc[row.name, 'DeltaSHAPE_in_cell'] = 1
        print('negative')
    else:
        df.loc[row.name, 'DeltaSHAPE_ex_vivo'] = 0
        df.loc[row.name, 'DeltaSHAPE_in_cell'] = 0
        print('neither')

print()
```

```python
file_list = glob.glob("../../Data/Bed files Xist/CLIPdb_transcript_coords/*.bed")
for i in file_list:
    print(i)
```

```python
proteins = set()
for i in file_list:
    file = os.path.basename(i)
    protein = file.split("_")[0]
    print(protein)
    proteins.add(protein)
    if protein not in df.columns:
        df[protein] = 0
    #else:
    data = pd.read_csv(i,sep="\t",header=None,index_col=None,names=["gene",\
                                                                        "start",\
                                                                        "end",\
                                                                        "name",\
                                                                        "coverage",\
                                                                        ".",\
                                                                        "p_value"])
    for j in data.index:
        start = int(data.loc[j,"start"])          
        end = int(data.loc[j,"end"])
        coordinates = np.arange(start,end+1)
        df.loc[coordinates,protein] = 1
```

```python
df.to_csv("../../Data/All_data_table_v2.txt",sep="\t")
```
