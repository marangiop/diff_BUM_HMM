```python
import pandas as pd
import numpy as np
import os
import glob
```

```python
colnames=['nucleotide_number', 'SHAPE_reactivity', 'SE_reactivity', 'nucleotide_identity'] 
df = pd.read_csv("../../Data/Map files Xist/XIST_exvivo_1M7_rep1.map", sep="\t",names=colnames, header=None,index_col=0)
df.head()
```

### don't need the SHAPE reactivity, only the delta shape values so the first two columns can go:

```python
df = df.drop(df[["SHAPE_reactivity","SE_reactivity"]],axis=1)
```

```python
#Create empty column
df["DeltaSHAPE_ex_vivo"] = 0
df["DeltaSHAPE_in_cell"] = 0
```

```python
df.head()
```

```python
#ADDING DELTA SHAPE INFORMATION

exvivo = pd.read_csv('../../Data/Map files Xist/XIST_exvivo_1M7.txt', sep="\t", header=0, index_col=0)
invivo = pd.read_csv('../../Data/Map files Xist/XIST_incell_1M7.txt', sep="\t", header=0, index_col=0)
```

```python
exvivo.head()
```

```python
invivo.head()
```

```python
df["DeltaSHAPE_ex_vivo"].loc[exvivo.index] = exvivo["DeltaSHAPE"]
df["DeltaSHAPE_in_cell"].loc[invivo.index] = invivo["DeltaSHAPE"]
```

### Check if it has been inserted correctly!!:

```python
df[df["DeltaSHAPE_in_cell"] != 0].head()
```

```python
file_list = glob.glob("../../Data/Bed files Xist/CLIPdb_transcript_coords/*.bed")
for i in file_list:
    print(i)
```

```python
proteins = set()
for i in file_list:
    file = os.path.basename(i)
    protein = file.split("_")[0]
    print(protein)
    proteins.add(protein)
    if protein not in df.columns:
        df[protein] = 0
    #else:
    data = pd.read_csv(i,sep="\t",header=None,index_col=None,names=["gene",\
                                                                        "start",\
                                                                        "end",\
                                                                        "name",\
                                                                        "coverage",\
                                                                        ".",\
                                                                        "p_value"])
    for j in data.index:
        start = int(data.loc[j,"start"])          
        end = int(data.loc[j,"end"])
        coordinates = np.arange(start,end+1)
        df.loc[coordinates,protein] = 1
```
