```python
#CELLS 1-3 ARE THE CORRECT WAY TO PLOT THE FIRST 700 NT OF THE MOLECULE BASED ON TOBY'S VERSION

import matplotlib.pyplot as plt
import pandas as pd
from platform import python_version

print(python_version())
```

```python
#read output files and edit for plotting in barplots
#RUN THIS TO PLOT DIFFBUM_HMM RESULTS
df = pd.read_csv('C:/Users/maran/Desktop/diff_BUM_HMM_Project/Github/diff_BUM_HMM/Xist_in vivo_vs_ex vivo_diff_BUM_HMM_analysed.txt', sep="\t", header=0)
df= df.rename_axis('nt').reset_index()
#df = df.iloc [:700]
df['filteredUM'] = [0 if x<0.95 else x for x in df['UM']]
df['filteredMU'] = [0 if x<0.95 else x for x in df['MU']]
#print df[['filteredUM','filteredMU']]
df['filteredUM']=df['filteredUM']*-1
df['UM']=df['UM']*-1
df
```

```python
# filtered for posterior probabilities of modification of >=0.95
fig,ax = plt.subplots()
plotfilteredUM = ax.bar(df['nt'], df['filteredUM'])
plotfilteredMU = ax.bar(df['nt'], df['filteredMU'])
ax.set_xlabel('Nucleotide position')
ax.set_ylabel('Posterior probabilities')
ax.set_title('Xist lncRNA ex vivo vs. in vivo - entire molecule')
#ax.spines['bottom'].set_position(('data', 0))
ax.legend((plotfilteredMU, plotfilteredUM), ('More reactive in vivo', 'More reactive ex vivo'), loc=1, fontsize = 5)
fig.savefig("diffBUM_HMM output - Xist lncRNA ex vivo vs. in vivo 0.95 threshold entire molecule.pdf", dpi= 300)
```
