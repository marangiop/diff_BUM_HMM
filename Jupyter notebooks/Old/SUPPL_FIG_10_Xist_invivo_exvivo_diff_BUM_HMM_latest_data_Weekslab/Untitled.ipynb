```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('../../Xist_in vivo_vs_ex vivo_new_data_october_reanalysed_diff_BUM_HMM_analysed.txt', sep="\t", header=0)
df= df.rename_axis('nt').reset_index()
#df = df.iloc [:700]
df['filteredUM'] = [0 if x<0.95 else x for x in df['UM']]
df['filteredMU'] = [0 if x<0.95 else x for x in df['MU']]
#print df[['filteredUM','filteredMU']]
df['filteredUM']=df['filteredUM']*-1
df['UM']=df['UM']*-1
df
```

```python
# filtered for posterior probabilities of modification of >=0.95
fig,ax = plt.subplots()
plotfilteredUM = ax.bar(df['nt'], df['filteredUM'])
plotfilteredMU = ax.bar(df['nt'], df['filteredMU'])
ax.set_xlabel('Nucleotide position')
ax.set_ylabel('Posterior probabilities')
ax.set_title('Xist lncRNA ex vivo vs. in vivo - entire molecule - new data')
#ax.spines['bottom'].set_position(('data', 0))
ax.legend((plotfilteredMU, plotfilteredUM), ('More reactive in vivo', 'More reactive ex vivo'), loc=1, fontsize = 5)
fig.savefig("diffBUM_HMM_output_Xist_lncRNA_exvivo_vs_invivo_0.95_threshold_entire_molecule_new_data_reanalysed_with_latest_data.pdf", dpi= 300)
```
